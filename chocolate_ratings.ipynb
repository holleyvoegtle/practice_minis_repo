```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Request url and print html
webpage = requests.get('https://s3.amazonaws.com/codecademy-content/courses/beautifulsoup/cacao/index.html')
soup = BeautifulSoup(webpage.content, 'html.parser')
print(soup)
```

```python
# command on soup object to get all the tags that contain the ratings.
ratings_data = soup.find_all(attrs={"class": "Rating"})
ratings = []
for rating in ratings_data[1:]:
  ratings.append(float(rating.string))
print (ratings)
```

```python
# Using matplotlib, create a histogram of the ratings value
plt.hist(ratings)
plt.show()
```

```python
# find tags on the webpage that contain the company names; loop through tags and add text from each tag to list
company_data = soup.select('.Company')
companies = []
for company in company_data[1:]:
  companies.append(company.string)
print(companies)
```

```python
# Create a DataFrame with Company and Ratings
dict = {
  'Company': companies,
  'Rating': ratings
}
df = pd.DataFrame.from_dict(dict)
df.head()
```

```python
# Use .groupby to group df by company and take the average of groups ratings; use .nlargest to get 10 best
avg_ratings = df.groupby('Company').Rating.mean()
ten_best = avg_ratings.nlargest(10)
print(ten_best)
```

```python
# Create a list that contains all the cocoa %, store as a float after stripping the % character
cocoa_data = soup.select('.CocoaPercent')
cocoa_percents = []
for cocoa_percent in cocoa_data[1:]:
  cocoa_percents.append(int(float(cocoa_percent.string[:-1])))
print(cocoa_percents)
```

```python
# add the cocoa percentage as a column in the df that has companies and rating in it
df['CocoaPercentage'] = cocoa_percents
df.head()
```

```python
# Make a scatterplot of ratings vs. percentage of cocoa
plt.cla()
plt.scatter(df.CocoaPercentage, df.Rating)
plt.show
```

```python
# Is there a correlation? use numpy commands 
z = np.polyfit(df.CocoaPercentage, df.Rating, 1)
line_function = np.poly1d(z)
plt.plot(df.CocoaPercentage, line_function(df.CocoaPercentage), "r--")
plt.show()
```
